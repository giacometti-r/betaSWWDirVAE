```python
import os
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from PIL import Image
import numpy as np

def load_images_from_folder(folder):
    images = []
    filenames = [f'epoch_{epoch}.png' for epoch in np.arange(0,2000,10)]
    for filename in filenames:
        img = Image.open(os.path.join(folder, filename))
        img_array = np.asarray(img)
        img.close()
        images.append(img_array)
        if filename == 'epoch_1990.png':
            img = Image.open(os.path.join(folder, 'epoch_1999.png'))
            img_array = np.asarray(img)
            img.close()
            images.append(img_array)
    return images

def update(frame):
    plt.imshow(images[frame])
    return [im]

cwd = os.getcwd()

folder = f'{cwd}/dirvae_hyper_tuning_early/reconstructed_images/'  # Update this path to your folder
images = load_images_from_folder(folder)

fig = plt.figure()
im = plt.imshow(images[0], animated=True)

ani = FuncAnimation(fig, update, interval=100, frames=len(images), blit=True)
ani.save('animation.mp4', writer='ffmpeg')  # or 'animation.gif' for GIF format

plt.show()
```
